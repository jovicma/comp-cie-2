```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parametros

sigma = 10.0
rho = 27.0
beta = 8.0 / 3.0

Tf = 30
h= 0.005

N = int(Tf/h)
```

```python
#Condição Inicial

t= [0]
x= [1.0]
y= [0.1]
z= [0.1]
```

```python
def f(cx,cy,cz):
  f1 = sigma*(cy-cx)
  f2 = cx*(rho-cz)-cy
  f3 = cx*cy-beta*cz
  return [f1,f2,f3]
```

```python
for n in range(N):
  fn = f(x[-1],y[-1],z[-1])
  x.append(x[-1]+h*fn[0])
  y.append(y[-1]+h*fn[1])
  z.append(z[-1]+h*fn[2])
  t.append(t[-1]+h)
```

```python
plt.plot(t,x,'b')
plt.plot(t,y,'r')
plt.plot(t,z,'g')
```
